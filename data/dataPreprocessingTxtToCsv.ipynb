```python
import csv
 
#txt파일을 읽고 txt파일과 csv 파일로 추출할 예정.
openFileTxt = open('psaA.txt', 'r')

writeFileCsv = open('psaaData.csv', 'w')
wr = csv.writer(writeFileCsv)
 
lines = openFileTxt.readlines()
 
cnt = 0 
sequence = ''
head = ''

for line in lines :
    line = line.replace(",",".")
    # 각 DNA의 첫번째 줄 : 헤더부분일 때 
    if(cnt == 0):
        head = line.strip('\n') + ','
        cnt += 1
        
    # 각 DNA의 서열들     
    elif(cnt > 0 and line != '\n'):
        cnt += 1
        sequence = sequence + line.strip('\n')
    
    #공백일 경우 줄변환 & 초기화
    elif(line == '\n'):
        data = head + sequence + '\n'
        cnt = 0
        sequence = ''
        head = ''
        writeFileCsv.write(data)
        print(data)
        
openFileTxt.close()
writeFileCsv.close()
```

```python
openFileCsv = open('psaaData.csv', 'r')
writeFileCsv = open('deleteline.csv', 'w')
wr = csv.writer(writeFileCsv)
 
lines = openFileCsv.readlines()
cnt = 1

for line in lines :
    if(len(line) <= 500):
        cnt + 1
        continue
     #공백일 경우 줄변환 & 초기화
    else :        
        writeFileCsv.write(line)
        print(cnt)
        print(line)
        cnt = cnt + 1
        
openFileCsv.close()
writeFileCsv.close()
```
